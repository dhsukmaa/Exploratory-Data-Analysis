```python
from wordcloud import WordCloud, STOPWORDS
count = df.Country.value_counts()
count

sns.set_style("ticks")

wordcloud = WordCloud(
                    width=1500,
                    height=1000, 
                    scale=1,
                    normalize_plurals=False,
                    repeat=False,
                    random_state=42,
                    background_color='black').generate_from_frequencies(frequencies = count)

plt.figure(figsize=(15,7))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.tight_layout(pad=3)
plt.show()
```

```python
def plot_barh(y, x, title=None, xlabel=None, ylabel=None,figsize=(10,10),style='fivethirtyeight', color='blue', show_number=False):
    plt.style.use(style)
    fig, ax = plt.subplots(figsize=figsize)
    
    plt.barh(y, x, height=0.50, color=color)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    
    if show_number:
        for i, z in enumerate(x):
            ax.text(z, i - .25, str(z),)
    plt.show()
```

```python
plot_barh(df['Country'],df['LandArea'], 'LandArea  of each country', 'LandArea ', 'Country', figsize=(10,16) ,color='peru', show_number=True)
```

```python
plot_barh(df['Country'],df['MedAge'], 'MedAge  of each country', 'MedAge', 'Country', figsize=(10,16) ,color='indianred', show_number=True)
```

```python
migrasi = df[['Country','Migrants']][df['Migrants']>0]

plot_barh(migrasi['Country'], migrasi['Migrants'],title="Migrants > 0",
         xlabel='Migrants', ylabel='Countries', color='lightseagreen',show_number=True
         )
```

```python
milyar = 1000000000
overMilyar = df[['Country','Population']][df['Population'] > milyar]
title='Countries with over billion population'
plot_barh(y=overMilyar['Country'],x=overMilyar['Population'],color='orchid', title=title, xlabel='Population', figsize=(5,4))
plt.show()
```

```python
migran = 1
overOne = df[['Country','Migrants']][df['Migrants'] > migran]
title='Countries with over billion population'
plot_barh(y=overOne['Country'],x=overOne['Migrants'],color='orchid', title=title, xlabel='Migrants', figsize=(15,7))
plt.show()
```
