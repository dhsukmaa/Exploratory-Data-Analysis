```python
count_disc = atlet.Discipline.value_counts()
count_disc

sns.set_style("ticks")

wordcloud = WordCloud(
                    width=2000,
                    height=1000, 
                    scale=1,
                    normalize_plurals=False,
                    repeat=False,
                    random_state=42,
                    background_color='black').generate_from_frequencies(frequencies = count_disc)

plt.figure(figsize=(17,10))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.tight_layout(pad=2)
plt.show()

```

```python
count_noc = atlet.NOC.value_counts()
count_noc

sns.set_style("ticks")

wordcloud = WordCloud(
                    width=2000,
                    height=1000, 
                    scale=1,
                    normalize_plurals=False,
                    repeat=False,
                    random_state=42,
                    background_color='black').generate_from_frequencies(frequencies = count_noc)

plt.figure(figsize=(17,10))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.tight_layout(pad=2)
plt.show()

```
